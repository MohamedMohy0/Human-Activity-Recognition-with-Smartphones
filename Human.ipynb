```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, accuracy_score
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
```

```python
#check the nan in data
na=train.isnull().sum()*100/len(train)
na.sort_values(ascending=True)
```

```python
#split the data
y=train["Activity"]
train=train.drop(columns=['Activity'])
```

```python
#find corelation and draw heat map to check correlatin in data
cor=train.corr()
plt.figure(figsize=(100, 100))
sns.heatmap(cor, cmap='coolwarm')
plt.show()
```

```python
#remove columns that have high corrltaion
threshold = 0.75
cor = train.corr().abs()
upper = cor.where(np.triu(np.ones(cor.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > threshold)]
train = train.drop(columns=to_drop)
```

```python
#prepare test data for models
test_x=test.drop(columns=["Activity"])
test_y=test["Activity"]
l=list(train.columns)
test_x=test_x[l]
```

```python
#prepare models 
classifiers = {
    'Logistic Regression': LogisticRegression(max_iter=200),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'SVM': SVC(kernel='linear'),
    'Naive Bayes': GaussianNB(),
    'Neural Network': MLPClassifier(max_iter=200),
}

# Train and evaluate each classifier
for name, clf in classifiers.items():
    clf.fit(train, y)
    y_pred = clf.predict(test_x)
    print(f"{name} - Accuracy: {accuracy_score(test_y, y_pred)}")
    print(classification_report(test_y, y_pred))
    print('-'*80)
```
